5. What are iterators in Python and how do they differ from iterables?

Ans. Iterables :
- An iterable is any object that can be looped over using a for loop.
- It implements the __iter__() method.

✅ Examples:
Lists, Tuples, Strings, Dictionaries, Sets

```python
my_list = [1, 2, 3]
for i in my_list:
    print(i)
```

Iterators:
- An iterator is an object that remembers its position during iteration.
  It implements both:
- __iter__() → returns the iterator object itself
- __next__() → returns the next item, and raises 'StopIteration' when done
- You can create an iterator from an iterable using the iter() function.

```python
my_list = [1, 2, 3]
it = iter(my_list)

print(next(it))
print(next(it))
print(next(it))

```

6. Explain the concept of generators in Python and how they are defined?

Ans. A generator is a special type of iterator that yields values one at a time using the 'yield' keyword instead of 'return'.
Generators are used to:
- Generate large sequences without storing all values in memory
- Write memory-efficient and lazy-evaluation code

```python
# Using a Function with yield:
def countdown(n):
    while n > 0:
        yield n
        n -= 1
# call it:
gen = countdown(3)

print(next(gen))
print(next(gen))
print(next(gen))
```

7. What are the advantages of using generators over regular functions?

Ans.
- Generators don’t store all values in memory.
- They generate one value at a time using 'yield'.
- Values are produced only when needed, not all at once.
(Ideal for streaming data, big files, or real-time processing.)
- You can create functions that run forever without crashing your system.
- Generators avoid using temporary lists, reducing code complexity.

10. What is the difference between map(), reduce(), and filter() functions in Python?

Ans. 1. map()
Purpose: Applies a function to each item of an iterable and returns a new iterable with the results.

2. filter()
Purpose: Filters elements from an iterable by applying a function that returns True or False.

3. reduce()
Purpose: Applies a function cumulatively to the items of an iterable, reducing it to a single value.

```python
# Map
nums = [1, 2, 3]
result = map(lambda x: x * 2, nums)
print(list(result))  # [2, 4, 6]

# filter
nums = [1, 2, 3, 4]
result = filter(lambda x: x % 2 == 0, nums)
print(list(result))  # [2, 4]

# reduce
from functools import reduce
nums = [1, 2, 3, 4]
result = reduce(lambda x, y: x + y, nums)
print(result)  # 10
```

1. Write a Python function that takes a list of numbers as input and returns the sum of all even numbers in
the list.

```python
def sum_of_even_numbers(numbers):
    even_sum = 0
    for num in numbers:
        if num % 2 == 0:
            even_sum += num
    return even_sum


my_list = [1, 2, 3, 4, 5, 6, 7, 8]
result = sum_of_even_numbers(my_list)
print("Sum of even numbers:", result)
```

4. Write a Python function that checks if a given number is prime or not from 1 to 200.

```python
def is_prime(n):
    if n <= 1:
        return False
    for i in range(2, n):
        if n % i == 0:
            return False
    return True

# Example usage
num = int(input("Enter a number between 1 and 200: "))

if num > 200:
    print("Please enter a number within the range 1 to 200.")
elif is_prime(num):
    print(num, "is a prime number.")
else:
    print(num, "is not a prime number.")
```

5. Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of terms.

```python
class Fibonacci:
    def __init__(self, n):
        self.n = n          # How many numbers to generate
        self.count = 0      # How many numbers have been generated
        self.a = 0          # First number in Fibonacci sequence
        self.b = 1          # Second number in Fibonacci sequence

    def __iter__(self):
        return self         # The object itself is the iterator

    def __next__(self):
        if self.count >= self.n:
            raise StopIteration  # Stop when n numbers are generated

        if self.count == 0:
            self.count += 1
            return self.a
        elif self.count == 1:
            self.count += 1
            return self.b
        else:
            # Next number is sum of previous two
            next_num = self.a + self.b
            self.a = self.b
            self.b = next_num
            self.count += 1
            return next_num

# Using the iterator:
fib_sequence = Fibonacci(7)  # Generate first 7 Fibonacci numbers

for number in fib_sequence:
    print(number)
```

6. Write a generator function in Python that yields the powers of 2 up to a given exponent.

```python
def powers_of_two(n):
    for i in range(n + 1):
        yield 2 ** i

# Example usage:
for value in powers_of_two(5):
    print(value)
```

7. Implement a generator function that reads a file line by line and yields each line as a string.

```python
def read_lines(filename):
    file = open(filename, 'r')
    for line in file:
        yield line
    file.close()

# Usage example:
for line in read_lines('sample.txt'):
    print(line, end='')  # end='' to avoid double newlines

```

9. Write a Python program that uses `map()` to convert a list of temperatures from Celsius to Fahrenheit

```python
# Function to convert Celsius to Fahrenheit
def celsius_to_fahrenheit(c):
    return (c * 9/5) + 32

# List of temperatures in Celsius
celsius_temps = [0, 20, 37, 100]

# Use map() to convert all temperatures to Fahrenheit
fahrenheit_temps = list(map(celsius_to_fahrenheit, celsius_temps))

print("Temperatures in Fahrenheit:", fahrenheit_temps)
```

10. Create a Python program that uses `filter()` to remove all the vowels from a given string.

```python
def remove_vowels(input_str):
    vowels = "aeiouAEIOU"
    # filter out characters that are vowels
    result = filter(lambda char: char not in vowels, input_str)
    # join the filtered characters back to a string
    return ''.join(result)

# Example usage
text = "Hello, this is a sample string."
no_vowels_text = remove_vowels(text)
print(no_vowels_text)
```
